# Load the Data

```python
import pandas as pd

# Load the train and test data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Check the first few rows to make sure the data is loaded correctly
print(train.head())
print(test.head())
```

# Data Preparation

## 2.1 Extract Features and Labels
- Extract the label (target) from the train dataset and prepare the pixel data for both training and testing.

```python
# Extract labels from the training data
Y_train = train["label"]

# Drop the label column from X_train
X_train = train.drop(labels=["label"], axis=1)

# Check the dimensions of X_train and Y_train to confirm
print(X_train.shape)
print(Y_train.shape)
```

## 2.2 Check for Missing Values
- It's important to check for any missing values to ensure the dataset is clean.


```python
# Check for missing values in the training data
print(X_train.isnull().sum().sum())  # Should return 0 if no missing values

# Check for missing values in the test data
print(test.isnull().sum().sum())  # Should also return 0
```

# Normalize the Data

 - Normalize the pixel values, which currently range from 0 to 255, by dividing by 255 so they fall between 0 and 1.

```python
# Normalize the pixel values to be between 0 and 1
X_train = X_train / 255.0
test = test / 255.0
```

# Reshape the Data

- Reshape the 1D arrays of pixel data into 28x28 pixel matrices with 1 channel (grayscale)

```python
# Reshape the data to 28x28x1 (28x28 image size, 1 channel)
X_train = X_train.values.reshape(-1, 28, 28, 1)
test = test.values.reshape(-1, 28, 28, 1)

# Check the new shape
print(X_train.shape)
print(test.shape)
```

# Label Encoding

 - Convert the labels (0-9) into one-hot encoded vectors

```python
from tensorflow.keras.utils import to_categorical

# One-hot encode the labels
Y_train = to_categorical(Y_train, num_classes=10)

# Check the shape of Y_train after encoding
print(Y_train.shape)
```

# Split Training and Validation Sets

- Split the X_train and Y_train data into training and validation sets (90% training, 10% validation)

```python
from sklearn.model_selection import train_test_split

# Set a random seed for reproducibility
random_seed = 2

# Split the training data into 90% training and 10% validation
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.1, random_state=random_seed)

# Check the new shapes
print(X_train.shape, Y_train.shape)
print(X_val.shape, Y_val.shape)
```

# Building the CNN Model

 - Use Keras' Sequential API to build the CNN architecture. The architecture will include convolutional layers, max pooling, dropout (to prevent overfitting), and fully connected layers.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense, Input

# Initialize the CNN model
model = Sequential()

# Define the input layer explicitly
model.add(Input(shape=(28, 28, 1)))

# First block of Conv2D and MaxPooling
model.add(Conv2D(filters=32, kernel_size=(5,5), padding='Same', activation='relu'))
model.add(Conv2D(filters=32, kernel_size=(5,5), padding='Same', activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# Second block of Conv2D and MaxPooling
model.add(Conv2D(filters=64, kernel_size=(3,3), padding='Same', activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(3,3), padding='Same', activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# Flatten the data to feed into fully connected (Dense) layers
model.add(Flatten())

# Fully connected layer
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.5))

# Output layer (10 categories for digits 0-9)
model.add(Dense(10, activation="softmax"))
```

# Compile the Model

- Compile the model with an optimizer, loss function, and evaluation metric:

```python
from tensorflow.keras.optimizers import RMSprop

# Compile the model
optimizer = RMSprop(learning_rate=0.001)
model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=["accuracy"])
```

# Summary of the Model

- Take a look at the model summary to see the architecture

```python
# Print the model summary
model.summary()
```

# Train the Model

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

# Set a learning rate reduction callback to reduce the learning rate when accuracy plateaus
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', 
                                            patience=3, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.00001)

# Define the number of epochs and batch size
epochs = 10
batch_size = 86

# Train the model
history = model.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, 
                    validation_data=(X_val, Y_val), verbose=2, 
                    callbacks=[learning_rate_reduction])
```

#  Evaluate the Model

- Evaluate the model by plotting training and validation accuracy and loss

```python
import matplotlib.pyplot as plt

# Plot the loss and accuracy curves for training and validation
fig, ax = plt.subplots(2,1)
ax[0].plot(history.history['loss'], color='b', label="Training loss")
ax[0].plot(history.history['val_loss'], color='r', label="Validation loss")
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True)
plt.show()
```

# Analyze and Display Misclassified Images

```python
import numpy as np
import matplotlib.pyplot as plt

# Predict the values from the validation dataset
Y_val_pred = model.predict(X_val)

# Convert predictions classes to one hot vectors 
Y_val_pred_classes = np.argmax(Y_val_pred, axis=1)

# Convert validation observations to one hot vectors
Y_true = np.argmax(Y_val, axis=1)

# Errors are where the predicted label does not match the true label
errors = (Y_val_pred_classes != Y_true)

# Display some of the misclassified images
X_val_errors = X_val[errors]
Y_pred_classes_errors = Y_val_pred_classes[errors]
Y_true_errors = Y_true[errors]

def display_errors(errors_index, img_errors, pred_errors, obs_errors):
    """This function shows misclassified images with predicted and actual labels."""
    n = 0
    nrows = 2
    ncols = 3
    fig, ax = plt.subplots(nrows, ncols, sharex=True, sharey=True, figsize=(10,6))
    for row in range(nrows):
        for col in range(ncols):
            error = errors_index[n]
            ax[row, col].imshow((img_errors[error]).reshape((28, 28)), cmap='gray')
            ax[row, col].set_title(f"Pred: {pred_errors[error]}, True: {obs_errors[error]}")
            n += 1

# Get the indices of the errors
errors_index = np.argsort(np.max(Y_val_pred[errors], axis=1))[-6:]

# Display the top 6 misclassified images
display_errors(errors_index, X_val_errors, Y_pred_classes_errors, Y_true_errors)
```

# Display Correctly Classified Images

```python
# Correct predictions are where the predicted label matches the true label
correct = (Y_val_pred_classes == Y_true)

# Get the correctly classified images
X_val_correct = X_val[correct]
Y_pred_classes_correct = Y_val_pred_classes[correct]
Y_true_correct = Y_true[correct]

def display_correct_predictions(correct_index, img_correct, pred_correct, true_correct):
    """This function shows correctly classified images with predicted and actual labels."""
    n = 0
    nrows = 2
    ncols = 3
    fig, ax = plt.subplots(nrows, ncols, sharex=True, sharey=True, figsize=(10,6))
    for row in range(nrows):
        for col in range(ncols):
            correct_img = correct_index[n]
            ax[row, col].imshow((img_correct[correct_img]).reshape((28, 28)), cmap='gray')
            ax[row, col].set_title(f"Pred: {pred_correct[correct_img]}, True: {true_correct[correct_img]}")
            n += 1

# Get the indices of some correct predictions
correct_index = np.random.choice(np.arange(len(X_val_correct)), size=6, replace=False)

# Display the top 6 correctly classified images
display_correct_predictions(correct_index, X_val_correct, Y_pred_classes_correct, Y_true_correct)
```
